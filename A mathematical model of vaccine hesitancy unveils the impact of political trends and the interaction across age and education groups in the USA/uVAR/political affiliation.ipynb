```python
#PCA

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


df=pd.read_csv("parameter.csv")
data = df[["","","","A_r","R_r","Political_affiliation"]]
data["Political_affiliation"].replace({'R': "r", 'D': "b"}, inplace=True)
# Separate features (X) and labels (y)
X = data.drop('Political_affiliation', axis=1)  # Assuming 'label' is the name of your label column
y = data['Political_affiliation']
z=np.zeros([len(list(y))])


# Step 3: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Perform PCA
pca = PCA(n_components=2,svd_solver="randomized")
X_pca = pca.fit_transform(X_scaled)

# Step 5: Plot the data in 2D using the first two principal components
fig=plt.figure(figsize=(17,10))
rect_1=[0,0,.5,.5]
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, edgecolor='k', s=150)
plt.scatter([2],[0],c="r", edgecolor='k', s=150,label="Republican")
plt.scatter([-1],[0],c="b", edgecolor='k', s=150,label="Democratic")
plt.legend(loc=4,fontsize=22)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)

plt.xlabel("Principal component 1",fontsize=xlf)
plt.ylabel("Principal component 2",fontsize=xlf)
plt.savefig("PCA_state_wise.jpg", bbox_inches='tight', dpi=500)
```
